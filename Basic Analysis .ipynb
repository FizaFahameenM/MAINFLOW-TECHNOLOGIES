```python
import pandas as pd
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import MinMaxScaler
```

```python
# Step 1: Load the dataset
data = pd.read_csv('01.Data Cleaning and Preprocessing.csv')
```

```python
# Step 2: Handling Missing Data
# a. Identify missing values
print(data.isnull().sum())
```

```python
# b. Remove rows with missing values
data = data.dropna()
```

```python
# Step 3: Remove Duplicates
# a. Check for duplicate rows
print(data.duplicated().sum())
```

```python

# b. Remove duplicates
data = data.drop_duplicates()
```

```python
duplicate_count = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
data = data.drop_duplicates()
```

```python
import matplotlib.pyplot as plt

# List of columns to create box plots for
columns_to_plot = ['Y-Kappa', 'ChipRate', 'BF-CMratio', 'BlowFlow']

# Create a box plot for each column one by one
for column in columns_to_plot:
    plt.figure(figsize=(8, 6))
    plt.boxplot(data[column])
    plt.title(f'Box Plot for {column}')
    plt.xlabel(column)
    plt.ylabel('Values')
    plt.show()


    
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['Y-Kappa', 'ChipRate', 'BF-CMratio', 'BlowFlow']] = scaler.fit_transform(data[['Y-Kappa', 'ChipRate', 'BF-CMratio', 'BlowFlow']])
```

```python
scaler = MinMaxScaler()
for column in ['Y-Kappa', 'ChipRate', 'BF-CMratio', 'BlowFlow']:
    data[column] = scaler.fit_transform(data[column].values.reshape(-1, 1))
```

```python
# Step 6: Save the Cleaned Dataset
data.to_csv('cleaned_01.Data_Cleaning_and_Preprocessing.csv', index=False)
```
